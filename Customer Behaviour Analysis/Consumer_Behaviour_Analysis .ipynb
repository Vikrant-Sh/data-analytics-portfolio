```python
import numpy as np
import pandas as pd
```

```python
data= pd.read_csv('customer_shopping_behavior.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
#removing null vlaues from the data
data['Review Rating']= data.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
data.isnull().sum()
```

```python
#converting the column headers to snake case
data.columns =data.columns.str.lower()
```

```python
data.columns =data.columns.str.replace(' ','_')
```

```python
data
```

```python
data= data.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
data.columns
```

```python
# cereating new column
labels = ['Young Adult','Adult','Middle-Aged','Senior']
data['age_group'] = pd.qcut(data['age'],q=4,labels = labels)
```

```python
data[['age','age_group']]
```

```python
 # creating column purchase frequency

frequency = {'Fortnightly':14,'Weekly':7,'Monthly':30,'Quarterly':90,'Bi-Weekly':14,'Annually':365,"Every 3 Months":90}

data['purchase_frequency'] = data['frequency_of_purchases'].map(frequency)
```
